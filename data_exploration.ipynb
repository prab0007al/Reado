```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books = pd.read_csv(f"{path}/books.csv")
```

```python
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax= ax)
plt.xlabel("column")
plt.ylabel("Missing values")
plt.show()
```

```python
import numpy as np

books["missing description"] = np.where(books["description"].isna(), 1, 0);
books["age_of_books"] = 2025 - books["published_year"]
```

```python
columns_of_interest = ["num_pages", "age_of_books", "missing description", "average_rating"]
correlation_matrix = books[columns_of_interest].corr(method="spearman")

sns.set_theme(style="white")
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", cbar_kws={"label": "Spearman correlation"})
plt.title("Correlation Matrix")
plt.show()
```

we wanna know that is any of these columns have high  correlation with the missing description.
Reason we want to know this: if let say num_pages have higher correlation(closer to 1 or -1) with missing description that means either longer books or shorter books have missing description. and we will not recommend a book with missing description so we are maybe biased towards shorter or longer books.

As the correlation values are not that much we can delete the books which have missing description values.

```python
books[(books["description"].isna() )|
      (books["average_rating"].isna()) |
      (books["published_year"].isna() )|
      (books["num_pages"].isna())
]
```

so the number of missing values is 303 , so it well under the 5% of our dataset so it is safe to delete these rows

```python
# just opposite of above
books_missing = books[(~books["description"].isna() )&
      (~books["average_rating"].isna()) &
      (~books["published_year"].isna() )&
      (~books["num_pages"].isna())
]
books_missing
```

```python
books_missing_25_words = books_missing[books_missing["words_in_description"]>=25]
books_missing_25_words
```

```python
books_missing_25_words["title_and_subtitle"] = (
    np.where(books_missing_25_words["subtitle"].isna(), books_missing_25_words["title"],
             books_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis = 1))
)
```

```python
books_missing_25_words["tagged_description"] = books_missing_25_words[["isbn13", "description"]].astype(str).agg(" ".join, axis = 1);
```
